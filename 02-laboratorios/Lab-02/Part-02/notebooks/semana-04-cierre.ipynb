# Semana 4 - Cierre comparativo

Este notebook consolida la comparacion final con el paper y el resumen del laboratorio.

```python
from pathlib import Path
import pandas as pd
import subprocess
import sys

cwd = Path.cwd().resolve()
search_roots = [cwd, *cwd.parents]
PART02 = next((p for p in search_roots if p.name == 'Part-02' and p.parent.name == 'Lab-02'), None)
if PART02 is None:
    PART02 = next((p / 'laboratorios' / 'Lab-02' / 'Part-02' for p in search_roots if (p / 'laboratorios' / 'Lab-02' / 'Part-02').exists()), None)
if PART02 is None:
    raise FileNotFoundError('No se pudo ubicar Part-02')

METRICS_CSV = PART02 / 'evidencias' / 'base' / 'metricas-reproduccion.csv'
df = pd.read_csv(METRICS_CSV)
df
```

## Tabla final contra el paper

```python
paper = {
    'with_empty_values': {'mean': 2.0368, 'median': 1.8199},
    'with_empty_values_seed21': {'mean': 2.0368, 'median': 1.8199},
    'without_empty_values': {'mean': 2.0559, 'median': 1.7745},
}

rows = []
for _, row in df.iterrows():
    ref = paper[row['escenario']]
    rows.append({
        'escenario': row['escenario'],
        'paper_mean': ref['mean'],
        'reproduction_mean': row['mean_m'],
        'delta_mean': round(row['mean_m'] - ref['mean'], 4),
        'paper_median': ref['median'],
        'reproduction_median': row['median_m'],
        'delta_median': round(row['median_m'] - ref['median'], 4),
    })

comparison = pd.DataFrame(rows)
comparison
```

## Version compacta para exponer

```python
compact = comparison[['escenario', 'paper_mean', 'reproduction_mean', 'paper_median', 'reproduction_median']]
compact
```

## Grafico opcional

```python
try:
    import matplotlib.pyplot as plt
    plot_df = comparison.set_index('escenario')[['paper_mean', 'reproduction_mean']]
    ax = plot_df.plot(kind='bar', figsize=(9, 4), title='Comparacion de media con el paper')
    ax.set_ylabel('Error medio (m)')
    plt.tight_layout()
    plt.show()
except Exception as exc:
    print('No se pudo generar el grafico:', exc)
```
