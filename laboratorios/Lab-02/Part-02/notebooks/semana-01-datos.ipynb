# Semana 1 - Datos y Preparacion

Este notebook sirve para mostrar la preparacion del dataset y la generacion de fingerprints.

```python
from pathlib import Path
import json
import pandas as pd
import subprocess
import sys

cwd = Path.cwd().resolve()
search_roots = [cwd, *cwd.parents]
PART02 = next((p for p in search_roots if p.name == 'Part-02' and p.parent.name == 'Lab-02'), None)
if PART02 is None:
    PART02 = next((p / 'laboratorios' / 'Lab-02' / 'Part-02' for p in search_roots if (p / 'laboratorios' / 'Lab-02' / 'Part-02').exists()), None)
if PART02 is None:
    raise FileNotFoundError('No se pudo ubicar Part-02')

RAW = PART02 / 'datos' / 'raw' / 'Position-Annotated-BLE-RSSI-Dataset'
SUMMARY = PART02 / 'datos' / 'processed' / 'online' / 'lab02_ble_ferrero' / 'preparation_summary.json'
print('PART02:', PART02)
print('RAW   :', RAW)
print('SUMMARY:', SUMMARY)
```

## Verificacion del dataset

```python
checks = {
    'raw_dataset': RAW.exists(),
    'hst': (RAW / 'hst').exists(),
    'trk': (RAW / 'trk').exists(),
    'cnf': (RAW / 'cnf').exists(),
}
pd.DataFrame([{'item': k, 'exists': v} for k, v in checks.items()])
```

## Resumen de preparacion

```python
summary = json.loads(SUMMARY.read_text(encoding='utf-8'))
summary
```

```python
variants = []
for name, payload in summary['variants'].items():
    variants.append({
        'variant': name,
        'min_valid_sensors': payload['min_valid_sensors'],
        'offline_rows': payload['offline_rows'],
        'online_rows': payload['online_rows'],
        'offline_path': payload['offline_path'],
        'online_path': payload['online_path'],
    })
pd.DataFrame(variants)
```
