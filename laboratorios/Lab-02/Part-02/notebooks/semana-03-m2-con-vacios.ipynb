# Semana 3 - M2 con valores vacios

Este notebook compara la corrida base con el barrido de semillas del escenario `with_empty_values`.

```python
from pathlib import Path
import json
import pandas as pd
import subprocess
import sys

cwd = Path.cwd().resolve()
search_roots = [cwd, *cwd.parents]
PART02 = next((p for p in search_roots if p.name == 'Part-02' and p.parent.name == 'Lab-02'), None)
if PART02 is None:
    PART02 = next((p / 'laboratorios' / 'Lab-02' / 'Part-02' for p in search_roots if (p / 'laboratorios' / 'Lab-02' / 'Part-02').exists()), None)
if PART02 is None:
    raise FileNotFoundError('No se pudo ubicar Part-02')

BASE = PART02 / 'resultados' / 'comparativas' / 'metricas' / 'M2' / 'base' / 'with_empty_values_evaluation.json'
SEED_DIR = PART02 / 'resultados' / 'comparativas' / 'metricas' / 'M2' / 'seed_sweep' / 'with_empty_values'
print(BASE)
print(SEED_DIR)
```

## Corridas disponibles

```python
rows = []
paper_mean = 2.0368
paper_median = 1.8199

base_eval = json.loads(BASE.read_text(encoding='utf-8'))
base_stats = base_eval['distance_stats']
rows.append({
    'run': 'base',
    'mean': round(base_stats['mean'], 4),
    'median': round(base_stats['median'], 4),
    'delta_mean_vs_paper': round(base_stats['mean'] - paper_mean, 4),
    'delta_median_vs_paper': round(base_stats['median'] - paper_median, 4),
})

for path in sorted(SEED_DIR.glob('*_evaluation.json')):
    payload = json.loads(path.read_text(encoding='utf-8'))
    stats = payload['distance_stats']
    rows.append({
        'run': path.stem.replace('_evaluation', ''),
        'mean': round(stats['mean'], 4),
        'median': round(stats['median'], 4),
        'delta_mean_vs_paper': round(stats['mean'] - paper_mean, 4),
        'delta_median_vs_paper': round(stats['median'] - paper_median, 4),
    })

df = pd.DataFrame(rows).sort_values('mean')
df
```

## Mejor corrida

```python
best = df.iloc[0]
best
```
