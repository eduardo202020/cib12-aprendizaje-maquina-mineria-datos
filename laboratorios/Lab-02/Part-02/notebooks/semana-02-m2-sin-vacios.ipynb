# Semana 2 - M2 sin valores vacios

Este notebook resume la reproduccion estable del escenario `without_empty_values`.

```python
from pathlib import Path
import json
import pandas as pd
import subprocess
import sys

cwd = Path.cwd().resolve()
search_roots = [cwd, *cwd.parents]
PART02 = next((p for p in search_roots if p.name == 'Part-02' and p.parent.name == 'Lab-02'), None)
if PART02 is None:
    PART02 = next((p / 'laboratorios' / 'Lab-02' / 'Part-02' for p in search_roots if (p / 'laboratorios' / 'Lab-02' / 'Part-02').exists()), None)
if PART02 is None:
    raise FileNotFoundError('No se pudo ubicar Part-02')

TRAIN = PART02 / 'resultados' / 'experimentos' / 'M2' / 'base' / 'without_empty_values' / 'training_summary.json'
EVAL = PART02 / 'resultados' / 'comparativas' / 'metricas' / 'M2' / 'base' / 'without_empty_values_evaluation.json'
print(TRAIN)
print(EVAL)
```

## Resumen de entrenamiento

```python
training = json.loads(TRAIN.read_text(encoding='utf-8'))
training
```

## Metricas de evaluacion

```python
evaluation = json.loads(EVAL.read_text(encoding='utf-8'))
stats = evaluation['distance_stats']
pd.DataFrame([stats])
```

## Comparacion con el paper

```python
paper = {'mean': 2.0559, 'median': 1.7745}
comparison = pd.DataFrame([{
    'paper_mean': paper['mean'],
    'reproduction_mean': round(stats['mean'], 4),
    'delta_mean': round(stats['mean'] - paper['mean'], 4),
    'paper_median': paper['median'],
    'reproduction_median': round(stats['median'], 4),
    'delta_median': round(stats['median'] - paper['median'], 4),
}])
comparison
```
